```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn.linear_model
from sklearn.externals import joblib
import scipy
from scipy.stats.stats import pearsonr 
```

```python
tiles = ['t1','t2','t3','t4','t5','t6','t7','t8','t9','t10','t11','t12','t13','t14','t15','t16','t17','t18',
         't19','t20','t21','t22','t23','t24','t25','t26','t27','t28','t29','t30','t31']

#load in the base set of all data
tile_counts = pd.read_pickle('/u/home/m/mudiyang/scratch/Scaleup_counts_sequences/tile_scores.pkl')

base_input = pd.read_pickle('/u/home/m/mudiyang/scratch/Tile_Prediction_Regressors/Combined/data/base.pkl')
base_input_all = base_input.dropna()

#remove stuff used in combined models
for i in tiles:
    base_input_all = base_input_all.drop(columns=['actual_%s'%i])
```

```python
#initialize predictions DF
predictions = pd.DataFrame(index = base_input_all.index)

for tile in tiles:
    model = joblib.load('/u/home/m/mudiyang/scratch/Tile_Prediction_Regressors/Megan_Model/Regressors/megan_%s.pkl'
                        %tile)
    
    to_predict = base_input_all[tile]
    predicted = model.predict(to_predict)
    predictions[tile] = predicted
```
